```python
from bs4 import BeautifulSoup # parsing the HTML contents
import requests # requests for approval for the data to be web scrapped, and also to get the html contents
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Downloading contents of the web page
url = "https://www.worlddata.info/africa/nigeria/inflation-rates.php"
r = requests.get(url)
#r.status_code
soup = BeautifulSoup(r.content)
soup
# r.content
```

```python
list_row = []
for header in soup.find_all("th"):
  list_row.append(header.text)

df = pd.DataFrame(columns= list_row)
df
```

```python
tr_ = soup.find_all("tr")
list_of_rows = []
for table_row in range(1,len(tr_)):
    list_for_each_row = [] 
    tr_mini = tr_[table_row]
    for row in tr_mini:
        list_for_each_row.append(row.text)
    list_of_rows.append(list_for_each_row)

for i in list_of_rows:
    print(i)
    df.loc[len(df)] = i
    


#     list_of_rows.append(list_for_each_row)
#     print(list_of_rows)
#df1
# tr_mini
# list_for_each_row
# print(list_of_rows)
df
```

#Cleaning the dataset


*   investigating the data to check for missing values and anyother cleaning necessary
*   preparing the data for visualisation



```python
#reading the dataframe
df = pd.read_csv('/content/drive/MyDrive/STUTERN 1.3 ASSIGNMENT/INTERMEDIATE/CBN.csv')
df.head()
```

```python
df.info()
```

```python
#dropping irrelevant column
to_drop = ['Unnamed: 0']
 
df.drop(to_drop, inplace=True, axis=1)
```

```python
df.head()
```

```python
df.columns=df.columns.str.replace('[%,Ø]','')
df.head()
```

```python
df.rename(columns={'Ø EU': 'EU'},inplace=True)
df.rename(columns={'Ø USA': 'USA'},inplace=True)
df.rename(columns={'Ø World': 'World'},inplace=True)
df.head()
```

```python
#removing trailing whitespaces from columns
df.columns = df.columns.str.strip()
df.head()
```

```python
#Check for missing values
df.isnull().sum()
```

```python
df.shape
```

```python
#dropping missing values
df.dropna(inplace=True)
print(df)
```

```python
df.isnull().sum()
```

```python
df.head()
```

removing symbols from the dataframe

```python
df["EU"]= df["EU"].str.replace("%", "")
```

```python
df["Nigeria"]= df["Nigeria"].str.replace("%", "")
df["USA"]= df["USA"].str.replace("%", "")
df["World"]= df["World"].str.replace("%", "")
```
